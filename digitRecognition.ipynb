### 01. Finding the Problem - Application
 Handwritten Digit recognition


### 02.Collecting the data
 * Input: Pixel value of Input Image 
 * Output: 10 Digits (0 to 9)

### 03.Load and Summarize the Dataset

```python
from sklearn.datasets import load_digits
dataset = load_digits()
dataset.images.shape
dataset.target.shape
dataset.target
dataset.images[0]
dataiamgeLength=len(dataset.images)
```

```python
#Visulaization of the dataset
n=9 #no of sample out   of samples total 1797
import matplotlib.pyplot as plt
plt.gray()
plt.matshow(dataset.images[n])
plt.show()
dataset.images[n]
```

### 04.Segregating dataset into X and Y

```python
X=dataset.images.reshape((dataiamgeLength,-1))
y=dataset.target
```

### 05. Splitting Dataset to Train & Test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

### 06.Model Training and Evaluation

```python
from sklearn.svm import SVC
model_svc=SVC()
model_svc.fit(X_train,y_train)
```

```python
y_pred=model_svc.predict(X_test)
from sklearn.metrics import accuracy_score
print(f"Accuracy is :{accuracy_score(y_test,y_pred)*100:.2f}%")
```

```python
model_svc.get_params()
```

```python
#Prediction
n=15
result = model_svc.predict(dataset.images[n].reshape((1,-1)))
plt.imshow(dataset.images[n], cmap=plt.cm.gray_r, interpolation='nearest')     
print(result)
print("\n")
plt.axis('off')
plt.title('%i' %result)
plt.show()
```

```python
# Using 'ovr' (default)
svc_ovr = SVC(decision_function_shape='ovr')
svc_ovr.fit(X_train, y_train)
print(f"Accuracy (ovr): {svc_ovr.score(X_test, y_test)}")

# Using 'ovo'
svc_ovo = SVC(decision_function_shape='ovo')
svc_ovo.fit(X_train, y_train)
print(f"Accuracy (ovo): {svc_ovo.score(X_test, y_test)}")
```
